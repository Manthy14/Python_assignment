```python
titles = pd.read_csv('titles.csv', index_col=None)
titles.head()
```

```python
cast = pd.read_csv('cast.csv',index_col=None)
cast.head()
```

# 1. What are the ten most common movie names of all time?

```python
# Group the DataFrame by the "title" column and count occurrences
common_movie_names = titles['title'].value_counts()

# Get the top ten most common movie names
top_10_common_movie_names = common_movie_names.head(10)

print("Top ten most common movie names of all time:")
print(top_10_common_movie_names)
```

# 2. Which three years of the 1930s saw the most films released?

```python
# Filter for movies released in the 1930s
movies_1930s = titles[(titles['year'] >= 1930) & (titles['year'] <= 1939)]

# Group the DataFrame by the "year" column and count the number of films in each year
film_counts_1930s = movies_1930s['year'].value_counts()

# Get the top three years with the most films released
top_3_years_1930s = film_counts_1930s.head(3)

print("Top three years of the 1930s with the most films released:")
print(top_3_years_1930s)
```

# 3. Plot the number of films that have been released each decade over the history of cinema 

```python
import matplotlib.pyplot as plt

# Extract the release year from the 'year' column and calculate the decade
titles['decade'] = (titles['year'] // 10) * 10

# Group the DataFrame by 'decade' and count the number of films in each decade
films_per_decade = titles['decade'].value_counts().sort_index()

# Create a bar chart to visualize the number of films per decade
plt.figure(figsize=(10, 6))
films_per_decade.plot(kind='bar', color='skyblue')
plt.title('Number of Films Released Each Decade')
plt.xlabel('Decade')
plt.ylabel('Number of Films')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 4. Plot the number of "Hamlet" films made each decade

```python
# Filter for films with the title "Hamlet"
hamlet_films = titles[titles['title'] == 'Hamlet']

# Extract the release year from the 'year' column and calculate the decade
hamlet_films['decade'] = (hamlet_films['year'] // 10) * 10

# Group the DataFrame by 'decade' and count the number of "Hamlet" films in each decade
hamlet_films_per_decade = hamlet_films['decade'].value_counts().sort_index()

# Create a bar chart to visualize the number of "Hamlet" films per decade
plt.figure(figsize=(10, 6))
hamlet_films_per_decade.plot(kind='bar', color='lightgreen')
plt.title('Number of "Hamlet" Films Released Each Decade')
plt.xlabel('Decade')
plt.ylabel('Number of "Hamlet" Films')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 15. Plot the n-values of the roles that Judi Dench has played over her career

```python
# Filter for rows where the actor's name is "Judi Dench"
judi_dench_roles = cast[cast['name'] == 'Judi Dench']

# Filter out rows where "n" is NaN
judi_dench_roles = judi_dench_roles.dropna(subset=['n'])

# Create a scatter plot to visualize the "n" values
plt.figure(figsize=(10, 6))
plt.scatter(judi_dench_roles['year'], judi_dench_roles['n'], marker='o', alpha=0.5)
plt.title('Judi Dench - "n" Values of Roles Over Her Career')
plt.xlabel('Year')
plt.ylabel('"n" Value')
plt.grid(True)
plt.show()
```

# 18. How many leading (n=1) roles were available to actors and how many to actressesin 1950s?

```python
# Filter for rows where the year is between 1950 and 1959 (inclusive)
roles_1950s = cast[(cast['year'] >= 1950) & (cast['year'] <= 1959)]

# Count the number of leading roles (n=1) for actors and actresses in the 1950s
leading_roles_actors_1950s = len(roles_1950s[(roles_1950s['n'] == 1) & (roles_1950s['type'] == 'actor')])
leading_roles_actresses_1950s = len(roles_1950s[(roles_1950s['n'] == 1) & (roles_1950s['type'] == 'actress')])

print("Number of leading roles (n=1) for actors in the 1950s:", leading_roles_actors_1950s)
print("Number of leading roles (n=1) for actresses in the 1950s:", leading_roles_actresses_1950s)
```
