# 1. Make a bar plot of months in which movies with "Christmas" in their title tend to be released in the USA

```python
# Filter for rows where the movie title contains "Christmas" and the country is "USA"
christmas_movies_usa = release[(release['title'].str.contains('Christmas', case=False, na=False)) & (release['country'] == 'USA')]

# Extract the month from the release date and count the number of releases for each month
monthly_counts = christmas_movies_usa['date'].str.extract(r'(\d{4}-\d{2})').groupby(0)[0].count()

# Sort the months in chronological order
monthly_counts = monthly_counts.reindex(pd.to_datetime(monthly_counts.index).strftime('%Y-%m'))

# Create a bar plot to visualize the data
plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='bar', color='skyblue')
plt.title('Monthly Release of Movies with "Christmas" in Title (USA)')
plt.xlabel('Month')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Make a bar plot of months in which movies whose titles start with "The Hobbit" are released in the USA

```python
# Filter for rows where the movie title starts with "The Hobbit" and the country is "USA"
hobbit_movies_usa = release[(release['title'].str.startswith('The Hobbit', na=False)) & (release['country'] == 'USA')]

# Extract the month from the release date and count the number of releases for each month
monthly_counts = hobbit_movies_usa['date'].str.extract(r'(\d{4}-\d{2})').groupby(0)[0].count()

# Sort the months in chronological order
monthly_counts = monthly_counts.reindex(pd.to_datetime(monthly_counts.index).strftime('%Y-%m'))

# Create a bar plot to visualize the data
plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='bar', color='lightgreen')
plt.title('Monthly Release of Movies Starting with "The Hobbit" (USA)')
plt.xlabel('Month')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 3. Make a bar of the day of the week on which movies with "Romance" in their title tend to be released in the USA

```python
# Filter for rows where the movie title contains "Romance," and the country is "USA"
romance_movies_usa = release[(release['title'].str.contains('Romance', case=False, na=False)) & (release['country'] == 'USA')]

# Extract the day of the week from the release date and count the number of releases for each day
day_of_week_counts = romance_movies_usa['date'].apply(lambda x: pd.to_datetime(x).strftime('%A')).value_counts()

# Sort the days of the week in the correct order
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
day_of_week_counts = day_of_week_counts.reindex(days_order)

# Create a bar plot to visualize the data
plt.figure(figsize=(10, 6))
day_of_week_counts.plot(kind='bar', color='pink')
plt.title('Day of the Week for Release of Movies with "Romance" in Title (USA)')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Releases')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 6. In which months do films with Judi Dench tend to be released in the USA

```python
# Filter the release dates DataFrame for Judi Dench movies in the USA
judi_dench_usa = release[(release['country'] == 'USA')]

# Extract the month from the release date and count the number of releases for each month
judi_dench_usa['release_month'] = pd.to_datetime(judi_dench_usa['date']).dt.month
monthly_counts = judi_dench_usa['release_month'].value_counts().sort_index()

# Create a bar plot to visualize the data
plt.figure(figsize=(10, 6))
monthly_counts.plot(kind='bar', color='skyblue')
plt.title('Judi Dench Movies Release Months in the USA')
plt.xlabel('Month')
plt.ylabel('Number of Releases')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
